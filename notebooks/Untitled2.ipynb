```python
import numpy as np

from keras.models import Sequential, Model
from keras.layers import BatchNormalization, Input

features = np.load('../downloads/features/Vg043D46E7Q.npy')
print(features.shape)
```

```python
input_features = Input(batch_shape=(2, 20, 4096,), name='features')
output = BatchNormalization(mode=0, axis=-1)(input_features)
model = Model(input=input_features, output=output)
model.compile(optimizer='rmsprop', loss='mse')
print(model.summary())
```

```python
batch = np.zeros((2,20,4096))
batch[0,:,:] = features[:20,:]
batch[1,:,:] = features[20:40,:]
```

```python
np.set_printoptions(threshold=np.nan)
print(batch[0,0,:])
```

```python
Y = model.predict_on_batch(batch)
print(Y[0,0,:])
```
