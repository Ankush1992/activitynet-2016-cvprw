```python
import re
import numpy as np

with open('../work/scripts/training/lstm_activity_classification/logs/experiment_3.log', 'r') as f:
    text = f.read()
    
search = re.findall(r'- loss: \d+\.\d+', text)
results = [s[8:] for s in search]
results = np.array(results)[np.arange(0, 40000, 400)]
epochs = np.arange(0, 100)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(num=None, figsize=(18, 18), dpi=100)
plt.plot(epochs, results)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.axis([0, 100, 0, 2])
plt.show()
```
